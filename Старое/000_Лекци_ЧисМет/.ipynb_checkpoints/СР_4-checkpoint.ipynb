# В способ
# Интерполяционный многочлен Лагранжа
$$P(x)=L_1(x)+L_2(x)+...+L_n(x)$$
$$L_k(x)=y_k*	
\prod\limits_{i = 1;i\neq k}^n \frac{(x-x_i)}{(x_k-x_i)}$$
$ (x_1,y_1);(x_2,y_2);...(x_n,y_n);$ - интерполяционные узлы <br/>


```julia
#Программа пример для лагранжа
X=[2 4 6 8];X|>display
Y=[4 50 192 478];Y|>display
#for i = 1:size(X)[2];X[i]|>print;print(" ");end;println("");
#for i = 1:size(Y)[2];Y[i]|>print;print(" ");end;println("");

function L_1(x);
    function xX(x,i,j);return (x-X[j])/(X[i]-X[j]);end;
    Y[1]*xX(x,1,2)*xX(x,1,3)*xX(x,1,4)
end;
function L_k(x,k);
    rez=Y[k];
    for i=1:size(X)[2];
        if i!=k;
            rez=rez*(x-X[i])/(X[k]-X[i]);
        end;
    end;
    return rez;
end;
function L(x);L_k(x,1)+L_k(x,2)+L_k(x,3)+L_k(x,4);end;
L(3)|>display
```

Теперь нам интерестно, что же мы такое построили

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#################################################################
_xlim=(-10,15)
_ylim=(-500, 500)

plot!(x->(L(x)), label = L"L(x)",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX",ylabel = "OX")
plot!(size=(400,200))
```
