# МетодПроловинногоДеления
Разбор первого варианта.<br/>
1)$2x^3-3x^2-12x-5=0$;<br/>
Шаг первый. Графически локализовать корни.

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-2.5,4)
_ylim=(-25, 25)
plot(title = "")
plot!(x->(x^3-0.2*x^2+0.5x^1+1.4), label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

Шаг второй, считаем по формуле...

```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoDelenia(
    _a::Core.Real
    ,_b::Core.Real
    ,F=(2*x^3-3*x^2-12x^1-5)
    ,e::Core.Real=0.01
)
    i=0;
    _e=false;
    while (!(_e)&&(i<100));
        _c=(_a+_b)/2;
        _sfa=_a|>F|>Sign;
        _sfc=_c|>F|>Sign;
        _sfb=_b|>F|>Sign;
        "$(lpad(_a,10));$(lpad(_c,10));$(lpad(_b,10));"|>print;
        ""|>println;
        if(_sfa*_sfc<0);_b=_c;end;
        if(_sfc*_sfb<0);_a=_c;end;
        i=i+1;
        _e=Base.abs(_a-_b)<2*e
    end;
end;
MetoDelenia(-1,2,x->(x^3+0.5*x^2+0.5x),0.00000000001)
```

МетодНьютона уточнить корни до заданной точности.<br/>


```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoDelenia(
    _xlast::Core.Real
    ,F=(x^3+3*x^2-24*x^1-10)
    ,F_=(3x^2+3*2*x-24)
    ,e::Core.Real=0.01
)
    i=0;
    _flag =false
    while ((!_flag)&&(i<10));
        i=i+1;
        _xlast|>println;
        F(_xlast)|>println;
        F_(_xlast)|>println;
        _xnext=_xlast-F(_xlast)/F_(_xlast)
        #F(_xlast)|>println;F_(_xlast)|>println;
        _flag=Base.abs(_xlast-_xnext)<e
        "$_xnext $_flag"|>println;
        _xlast=_xnext;
    end;
end;
#MetoDelenia(1,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001);
#""|>println;
#MetoDelenia(4,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001)
#""|>println;
#MetoDelenia(-2,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001)
MetoDelenia(10,x->(2*x^3-8*x^2-30*x+1),x->(6*x^2-16*x-30),0.1);
```

# МетодХорд
Шаг первый. Графически локализовать корни.<br/>
1)$2x^3-3x^2-12x-5=0$;<br/>

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-5,10)
_ylim=(-75,20)
plot(title = "")
plot!(x->(x^3-9*x^2+6*x^1+3), label = L"x^3-3*x^2+12x^1-12",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
#plot!(size=(400,200))
```

Методом половинного деления уточнить корни до заданной точности.<br/>


```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoXord(
    _xlastlast::Core.Real
    ,_xlast::Core.Real
    ,F=(x^3-3*x^2+2.5)
    ,e::Core.Real=0.01
)
    i=0;
    _flag =false
    while ((!_flag)&&(i<10));
        i=i+1;
        _xnext=_xlast-(_xlast-_xlastlast)*F(_xlast)/(F(_xlast)-F(_xlastlast));
        _flag=(Base.abs(_xlast-_xnext)<e);
        "$_xnext $_flag"|>println;
        _xlast=_xlastlast;
        _xlast=_xnext;
    end;
end;
MetoXord(-6,-5,x->(x^3-9*x^2+6*x^1+3),0.1);
```

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-5,10)
_ylim=(-150, 50)
plot(title = "")
plot!(x->(x^3-9*x^2+6*x+3),label = L"x^3-0.2x^2+0.5x+1.4",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(800,400))
```
