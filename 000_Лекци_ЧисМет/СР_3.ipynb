# МетодПроловинногоДеления
Разбор первого варианта.<br/>
1)$2x^3-3x^2-12x-5=0$;<br/>
Шаг первый. Графически локализовать корни.

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-2.5,4)
_ylim=(-25, 25)
plot(title = "")
plot!(x->(2*x^3-3*x^2-12x^1-5), label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

Шаг второй, считаем по формуле...

```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoDelenia(
    _a::Core.Real
    ,_b::Core.Real
    ,F=5-log(10,x)
    ,e::Core.Real=0.01
)
    i=0;
    _e=false;
    while (!(_e)&&(i<100));
        _c=(_a+_b)/2;
        _sfa=_a|>F|>Sign;
        _sfc=_c|>F|>Sign;
        _sfb=_b|>F|>Sign;
        "$(lpad(_a,10));$(lpad(_c,10));$(lpad(_b,10));"|>print;
        ""|>println;
        if(_sfa*_sfc<0);_b=_c;end;
        if(_sfc*_sfb<0);_a=_c;end;
        i=i+1;
        _e=Base.abs(_a-_b)<2*e
    end;
end;
MetoDelenia(-2,-1,x->(2*x^3-3*x^2-12x^1-5),0.0001)
```

# МетодНьютонаДеления
Шаг первый. Графически локализовать корни.<br/>
1)$2x^3-3x^2-12x-5=0$;<br/>

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-2.5,4)
_ylim=(-25, 25)
plot(title = "")
plot!(x->(2*x^3-3*x^2-12x^1-5), label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

Методом половинного деления уточнить корни до заданной точности.<br/>


```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoXord(
    _xlastlast::Core.Real
    ,_xlast::Core.Real
    ,F=(x^3+3*x^2-24*x^1-10)
    ,e::Core.Real=0.01
)
    i=0;
    _flag =false
    while ((!_flag)&&(i<10));
        i=i+1;
        _xnext=_xlast-(_xlast-_xlastlast)*F(_xlast)/(F(_xlast)-F(_xlastlast));
        _flag=(Base.abs(_xlast-_xnext)<e);
        "$_xnext $_flag"|>println;
        _xlast=_xlastlast;
        _xlast=_xnext;
    end;
end;
MetoXord(-3,-2,x->(x^3+3*x^2-24*x^1-10),0.000001);
```
