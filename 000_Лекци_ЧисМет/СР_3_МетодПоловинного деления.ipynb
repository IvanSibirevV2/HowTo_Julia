Методом половинного деления уточникть корни до заданной точности

```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoDelenia(
    _a::Core.Real
    ,_b::Core.Real
    ,F=5-log(10,x)
    ,e::Core.Real=0.01
)
    i=0;
    _e=false;
    while (!(_e)&&(i<100));
        _c=(_a+_b)/2;
        _sfa=_a|>F|>Sign;
        _sfc=_c|>F|>Sign;
        _sfb=_b|>F|>Sign;
        "$(lpad(_a,10));$(lpad(_c,10));$(lpad(_b,10));"|>print;
        ""|>println;
        if(_sfa*_sfc<0);_b=_c;end;
        if(_sfc*_sfb<0);_a=_c;end;
        i=i+1;
        _e=Base.abs(_a-_b)<2*e
    end;
end;
MetoDelenia(-1,0,x->(x^3+3*x^2-24*x^1-10),0.0001)
```
