### Теоретическая часть. Погрешность.
- x - точное значение изменяемой величины.
- а - приближенное значение изменяемой величины.
- $x-a$ - называют погрешностью числа а.
- $\Delta=|x-a|$ - **абсолютная погрешность числа $a$**

- Если точное значение измеряемой величины ($x$) неизвестно то тогда его берут равным приближенному значению измеряемой величины ($a$). ($x=a$)
- $\Delta_a$ - **предельная абсолютная погрешность**, определяется как минимальная погрешность большая или равная абсолютной погрешности.
- $\Delta_a=min \Delta : |x-a| \leq \Delta_a \leq \Delta \leq \Delta \leq ... \forall \Delta$
- **Относительной погрешностью** $\delta_a$ числа $a$ называют его отношение к его предельной абсолютной погрешности к точному значению числа $x$ (или приближенному значению числа $a$). $$\delta=\frac{\Delta}{x}; \delta=\frac{\Delta}{a};$$
- **Предельной относительной погрешностью** $\delta_a$ числа $a$ называют его отношение к его предельной абсолютной погрешности к точному значению числа $x$ (или приближенному значению числа $а$)
$$\delta_a=\frac{\Delta_a}{x};\delta_a=\frac{\Delta_a}{a};$$

$$|x-a|\leq\Delta_a$$
$$-\Delta_a\leq x-a \leq +\Delta_a$$
$$-\Delta_a+a\leq x-a+a \leq +\Delta_a+a$$
$$a-\Delta_a\leq x \leq a+\Delta_a$$
**Так и появилась следующая запись**
$$x=a\pm\Delta_a$$

# 1.3 Действия с приближенными числами
**Абсолютная погрешность алгеброической суммы** ($\Delta_{a\pm b}$) нескольких приближенных чисел ($a$ и $b$) не превышаетсуммы абсолютных погрешностей этих чисел ($\Delta_a + \Delta_b$ ). $$\Delta_{a\pm b}\leq\Delta_a + \Delta_b$$

### Правило сложения и вычитания приближенных чисел:
- Шаг первый: Выделить наименее точное число (или числа), то есть такое, в десятичной записи которого наименьшее число верных десятичных знаков;
- Шаг второй: Округлить остальные числа так, чтобы каждое из них содержало на один (запасной) знак больше, чем выделенное число;
- Шаг третий: Выполнить сложение и вычитание с учетом сохраненных знаков;
- Шаг четвертый: Полученный результат округлить до предпоследнего знака.
  - Правила округления числа
  - А) Если первая из отбрасываемых цифр меньше 5, то сохраняемые десятичные знаки остаются без изменения.
  - Б) Если первая из отбрасываемых цифр больше 5, то последний из сохраняемых знаков увеличивают на 1.
  - В) Если первая из отбрасываемых цифр равна 5, а среди следующих за ней цифр есть отличные от нуля, то последний из сохраняемых знаков увеличивают на 1
  - Г) Если первая из отбрасываемых цифр равна 5, а все последующие - нули, то последний из сохраняемых десятичных знаков увеличивают на 1, когда он нечетен, и сохраняют неизменным, когда он четен (правило четной цифры).

### Теперь даваяте попробуем сложить числа.
$a=414.8$ ; $b=0.025$ ; $c=24.17$ ; $d=0.000326$ ;
Наименее точное слагаемое можно определить либо через относительную погрешность, либо по минимальному колличеству значащих знаков после запятой... Округлим согластно правилу опираяся на количество знаков наименее точного числа.

**Относительная погрешность произведения (частного)** ($\delta_{\frac{ab}c}$) приближенных чисел не превышает суммы относительных погрешностей этих чисел ($\delta_a+\delta_b+\delta_c$). $$\delta_{\frac{ab}c}\leq\delta_a+\delta_b+\delta_c$$
### Правило сложения и вычитания приближенных чисел:
- Шаг первый: из всех чисел, которые предстоит умножать и делить, выделить наименее точное - то, в котором меньше всего верных значащих цифр;
- Шаг второй: *округлить* остальные числа так, чтобы каждое из них содержало на одну (запасную) значащую цифру больше, чем выделенное число;
- Шаг третий: выполнить умножение и деление округленных чисел с учетом сохраненных значащих цифр;
- Шаг четвертый: оставить в ответе столько значащих цифр, сколько их было в наименее точном числе.

Теперь нужно получить **предельную абсолютную погрешность**
$$\Delta=0.0003\leq0.0004=\Delta_a$$
Это единственный способ получить придельную абсолютную погрешность не добавляя знаков после запятой...
Теперь найдем **предельную относительную погрешность** $\delta_a=\frac{\Delta_a}{|a|}$

```julia
using Latexify,LaTeXStrings,SymEngine
_Local_Delta_A=0.0004
_Local_a=0.474
_Local_resalt=100*_Local_Delta_A/_Local_a
_Local_resalt2=Base.round(_Local_resalt; digits=2);
L"\delta_a=\frac{\Delta_a}{|a|}=\delta_a=100*\frac{%$_Local_Delta_A}{|%$_Local_a|}=%$_Local_resalt \%=%$_Local_resalt2\%"
```
