```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
```

```julia
function MetodIterachii(
    _Xlast::Core.Real
    ,F=5-log(10,x)
    ,e::Core.Real=0.01
)
    _Xnext=0;
    _flag=false;
    i=1;
    while !_flag;
        _Xnext=F(_Xlast);
        _abs=abs(_Xnext-_Xlast);
        _flag=e>abs(_Xnext-_Xlast);
        L"x_{%$i}=%$_Xnext;%$_flag==%$e>%$_abs"|>display
        _Xlast=_Xnext;i=i+1;
        if i>10; _flag=true end;
    end;
end;
MetodIterachii(1,x->(2*x^3-3*x^2-12x^1-5)/10,0.0001)
#MetodIterachii(2,x->((10-cos(x))/11),0.01)
```

```julia
_xlim=(-1.5,-1.25)
_ylim=(-1, 1)
plot(title = "Первый корень")
plot!(x->(2*x^3-3*x^2-12x^1-5)/10, label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

```julia
function F(x)
    return 2*x^3-3*x^2-12x^1-5;
end;
function FF(x)
    return 2*3*x^2-2*3*x-12;
end;
FF(-1.45)/10|>display
```
