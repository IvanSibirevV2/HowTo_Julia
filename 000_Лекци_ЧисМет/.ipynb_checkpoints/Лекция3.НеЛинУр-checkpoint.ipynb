# –¢–µ–º–∞ 3. –†–µ—à–µ–Ω–∏–µ –Ω–µ–ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(x)=0$. –ö–æ—Ä–Ω–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏—è –Ω–∞–∑—ã–≤–∞—é—Ç —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–∏ –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–µ –≤ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–±—Ä–∞—â–∞–µ—Ç –µ–≥–æ –≤ –≤–µ—Ä–Ω–æ–µ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ. <br/>
–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–∂–µ—Ç : –∏–º–µ—Ç—å –æ–¥–∏–Ω –∫–æ—Ä–µ–Ω—å $2x+1=0$; <br/>
–∏–º–µ—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –∫–æ—Ä–Ω–µ–π $x^2-5x+6=0$; <br/>
–∏–º–µ—Ç—å —Å–∫–æ–ª—å —É–≥–æ–¥–Ω–æ –º–Ω–æ–≥–æ –∫–æ—Ä–Ω–µ–π $\sin{x}=\frac{1}{2}$; <br/>
–¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã—Ö –∫–æ—Ä–Ω–µ–π –Ω–µ –∏–º–µ—Ç—å –≤–æ–≤—Å–µ $\sin{x}=\frac{1}{2}$; <br/>
–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(x)=0$ –º–æ–∂–Ω–æ —Ä–µ—à–∞—Ç—å –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏,  –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—É—Ç–µ–º –≤—ã—Ä–∞–∂–µ–Ω–∏—è –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π. –ù–æ —ç—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ –Ω–µ –≤—Å–µ–≥–¥–∞. –¢–æ–≥–¥–∞ –Ω–∞ –ø–æ–º–æ—â—å –ø—Ä–∏—Ö–æ–¥–∏—Ç —Å–ø–æ—Å–æ–± –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –∏–ª–∏ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —á–∏—Å–ª–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤.<br/>

# –ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π —Å–ø–æ—Å–æ–±
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(ùë•)=0$. –ö–æ—Ä–Ω–∏ —ç—Ç–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ª–µ–∂–∞—Ç –Ω–∞ –ø–µ—Ä–µ—á–µ—á–µ–Ω–∏–∏ $f(ùë•)=y$ –∏ $y=0$. <br/>

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#################################################################
_xlim=(-2.5,2.5);_ylim=(-2, 2);
plot(title = "=(2*x+1)=0")
plot!(x->(2*x+1), label = "y=(2*x+1)",xlim=_xlim,ylim=_ylim)
plot!(x->0, label = "y=0", xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX",ylabel = "OX")
plot!(size=(400,200))
```

–£ –Ω–∞—Å –Ω–µ –≤—Å–µ–≥–¥–∞ –µ—Å—Ç—å –¥–æ—Å—Ç—É–ø –∫ –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ–π —Ç–µ—Ö–Ω–∏–∫–µ, –ø–æ —ç—Ç–æ–º—É –∏–Ω–æ–≥–¥–∞ –∏—Å–ø–æ–ª—å–∑—É—é—Ç —Å–ª–µ–¥—É—é—â–∏–π —Å–ø–æ—Å–æ–± –ª–æ–∫–∞–ª–∏—Ö–∞—Ü–∏–∏ –∫–æ—Ä–Ω–µ–π. <br/>
–°—Ç—Ä–æ—è—Ç —Ç–∞–±–ª–∏—Ü—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç $f(x)=y$

```julia
Arr =  Array{NamedTuple{(:x, :y),Tuple{Int64,Int64}}}(undef, 0)
#pop!(Arr)
function f(x); return 2x+1; end;
push!(Arr,(x=-2,y=f(-2)))
push!(Arr,(x=-1,y=f(-1)))
push!(Arr,(x=0,y=f(0)))
push!(Arr,(x=1,y=f(1)))
push!(Arr,(x=2,y=f(2)))
Arr|>display
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ –ø—Ä–∏–º–µ—Ä–∞, —Å–º–µ–Ω–∞ –∑–Ω–∞–∫–æ–≤ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[-1,0]$. <br/>
–¢–æ–≥–¥–∞ –≥–æ–≤–æ—Ä—è—Ç —á—Ç–æ, –∫–æ—Ä–µ–Ω—å –ª–æ–∫–∞–ª–∏–∑–æ–≤–∞–Ω –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[-1,0]$. <br/>
–ú–∞ –¥–∞–∂–µ –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å –Ω–∞ –≤—Å–∫–∏–¥–∫—É —á—Ç–æ –∫–æ—Ä–µ–Ω—å —Ä–∞–≤–µ–Ω 0.5 . –û—Ç–≤–µ—Ç –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–Ω—ã–º.

–í–æ–∑–º–æ–∂–µ–Ω —Å–ª—É—á–∞–π –∫–æ–≥–¥–∞ —Ñ—É–Ω–∫—Ü–∏—è –∏–º–µ–µ—Ç –≤–∏–¥ $f(x)=g(x)$.<br/>
–¢–æ–≥–¥–∞ –º–æ–∂–Ω–æ —Å–≤–µ—Å—Ç–∏ –∑–∞–¥–∞—á—É –∫ –ø—Ä–µ–¥—ã–¥—É—â–µ–π:$f(x)-g(x)=0$.<br/>
–ò–ª–∏ –∏—Å–∫–∞—Ç—å –∫–æ—Ä–µ–Ω—å –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ñ—É–Ω–∫—Ü–∏–π.

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#################################################################
_xlim=(-2.5,2.5);_ylim=(-2, 2);
plot(title = "(2*x+1)=(1-3*x)")
plot!(x->(2*x+1), label = "y=(2*x+1)",xlim=_xlim,ylim=_ylim)
plot!(x->(1-3*x), label = "y=(1-3*x)",xlim=_xlim,ylim=_ylim)
plot!(x->0, label = "y=0", xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX",ylabel = "OX")
plot!(size=(400,200))
```

–ì—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ –ø—Ä–∏–≤–ª–µ–∫–∞–µ—Ç —Å–≤–æ–µ–π –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç—å—é, –Ω–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–Ω—ã–º, —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —á–µ–ª–æ–≤–µ—á–µ—Å–∫–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∞. –û—Ü–µ–Ω–∫–∞ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å —Ç–∞–∫–∂–µ –æ—Å—Ç–∞–µ—Ç—Å—è –∑–∞ –±–æ—Ä—Ç–æ–º.<br/>
–ü–æ —ç—Ç–æ–º—É –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –º–µ—Ç–æ–¥ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Ç–æ—á–∫–∏ —Å—Ç–∞—Ä—Ç–∞ –∏ —É–º–µ–Ω—å—à–µ–Ω–∏—è –∫–æ–ª-–≤–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π —á–∏—Å–ª–µ–Ω–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤.

# –ú–µ—Ç–æ–¥ –ø–æ–ª–æ–≤–∏–Ω–Ω–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è
–¢–µ–æ—Ä–µ–º–∞: e—Å–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è $y=f(x)$ <br/>
- –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[a_0;b_0]$; <br/>
- –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –∏–ª–∏ —É–±—ã–≤–∞–µ—Ç –Ω–∞ —ç—Ç–æ–º –æ—Ç—Ä–µ–∑–∫–µ; <br/>
- –Ω–∞ –∫–æ–Ω—Ü–∞—Ö –æ—Ç—Ä–µ–∑–∫–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞–∫–æ–≤, —Ç.–µ. $f(a)f(b)<0$; <br/>

–¢–æ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[a_0;b_0]$; –∏–º–µ–µ—Ç –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –∫–æ—Ä–µ–Ω—å. <br/>
–ï—Å–ª–∏ —É—Å–ª–æ–≤–∏—è –≤—ã–ø–æ–ª–Ω–µ–Ω–æ, —Ç–æ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å. –ö–∞–∂–¥–æ–µ —É—Å–ª–æ–≤–∏–µ —Ç–µ–æ—Ä–µ–º—ã –≤–∞–∂–Ω–æ!!!<br/>

**–ê–ª–≥–æ—Ä–∏—Ç–º –º–µ—Ç–æ–¥–∞ –ø–æ–ª–æ–≤–∏–Ω–Ω–æ–≥–æ –¥–µ–ª–µ–Ω–∏—è**
- 1) —Ä–∞–∑–±–∏—Ç—å –Ω–∞—á–∞–ª—å–Ω—ã–µ –æ—Ç—Ä–µ–∑–æ–∫ $[a_0;b_0]$ –ø–æ–ø–æ–ª–∞–º;
- 2) –ø–æ–ª—É—á–∏—Ç—å –∑–∞–ø–∏—Å—å –≤–∏–¥–∞
<table>
   <tr> <th>$a_0$</th> <th>$c_0$</th> <th>$b_0$</th> </tr>

–ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ –≤—ã–≥–ª—è–¥–∏—Ç —ç—Ç–æ –∫–∞–∫:<br/>

–ò–∑–Ω–∞—á–∞–ª—å–Ω–æ $sign(f(c_0))$ –¥–ª—è $c_0$ - –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ, –ø–æ —ç—Ç–æ–º—É –ø—Ä–∏–¥–µ—Ç—Å—è –µ–≥–æ —Ä–∞—Å—á–∏—Ç–∞—Ç—å.<br/>
–ü–æ—Å–ª–µ —Ä–∞—Å—á–µ—Ç–∞ $sign(f(c_0))=sign(f(\frac{a_0+b_0}{2}))$ –ø–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –∑–Ω–∞–∫...

   <tr> <th>[-2</th><th>-0.5]</th> <th>-1</th> </tr>

–ò –∏–∑ –¥–≤—É—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –≤—ã–±–∏—Ä–∞–µ–º —Ç–æ—Ç, –Ω–∞ –∫–æ–Ω—Ü–∞—Ö –∫–æ—Ç–æ—Ä–æ–≥–æ $f(x)$ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ä–∞–∑–Ω—ã–µ –∑–Ω–∞–∫–∏. –í—ã–±—Ä–∞–Ω –∏–Ω—Ç–µ—Ä–≤–∞–ª $[a_1;b_1]=[-2;-0.5]$. –î–ª—è –Ω–µ–≥–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ä–∞—Å—á–∏—Ç–∞—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏.<br/> 
- 3) –ö—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏.
$\varepsilon=0.1$ - –Ω–∞–ø–µ—Ä–µ–¥ –∑–∞–¥–∞–Ω–∞—è —Ç–æ—á–Ω–æ—Å—Ç—å;$2\varepsilon=0.2>|a_1-b_1|=1.5$;
–ö—Ä–∏—Ç–µ—Ä–∏–π –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç, —Å—á–∏—Ç–∞–µ–º –¥–∞–ª—å—à–µ
- 4) –ö–∞–∫ —Ç–æ–ª—å–∫–æ –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç, —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ä–∞—Å—á–∏—Ç–∞—Ç—å  –ø–æ –ø–æ–ª—É—á–µ–Ω–Ω–æ–º—É –∏–Ω—Ç–µ—Ä–≤–∞–ª—É –æ—Ç–≤–µ—Ç –∏ –µ–≥–æ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: <br/>
$X_n=\frac{a_n+b_n}{2}$ <br/>
–ü—Ä–µ–¥–µ–ª—å–Ω–∞—è –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: <br/>
$\Delta=\frac{|a_n-b_n|}{2}<\varepsilon$ <br/>

# –ú–µ—Ç–æ–¥ –ø—Ä–æ—Å—Ç–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–π
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(x)=0$. –ü—Ä–∏–≤–µ–¥–µ–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫ –≤–∏–¥—É $g(x)=x$.<br/>
$f(x)=0$;$f(x)+x=x$;$f(x)+x=g(x)$;$g(x)=x$;<br/>
–î–∞–ª–µ–µ –±—É–¥–µ–º —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–≤–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(x)=x$.<br/>
- –î–ª—è –Ω–∞–∏–±—ã—Å—Ç—Ä–µ–π—â–µ–π —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –ª–æ–∫–∞–ª–∏–∑–æ–≤–∞—Ç—å –∫–æ—Ä–Ω–∏ –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –∏–Ω—Ç–µ—Ä–∞–≤–ª–∞ $[a,b]$<br/>
- –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—É—é –∑–∞–¥–∞—á—É –Ω–∞ —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å. –ê–ª–≥–æ—Ä–∏—Ç–º —è–≤–ª—è–µ—Ç—Å—è —Å—Ö–æ–¥—è—â–∏–º—Å—è –µ—Å–ª–∏ $|f\prime(x)|<1$.<br/> –ü–æ–º–Ω–∏—Ç–µ –µ—Å–ª–∏ $|f\prime(x)|>1$, —Ç–æ –∞–ª–≥–æ—Ä–∏—Ç–º –±—É–¥–µ—Ç —Ä–∞—Å—Ö–æ–¥—è—â–∏–º—Å—è. –ü—Ä–æ–≤–µ—Ä—è–π—Ç–µ!!!<br/>

–ê–ª–≥–æ—Ä–∏—Ç–º –º–µ—Ç–æ–¥–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π.<br/>

- 1) –õ–æ–∫–∞–ª–∏–∑—É–µ–º –∫–æ—Ä–Ω–∏ –Ω–∞ –∏–Ω—Ç–µ—Ä–∞–≤–∞–ª–µ $[a,b]$
- 2) –í—ã–±–µ—Ä–∞–µ–º —Ç–æ—á–∫—É $x_0$ –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–æ $|f\prime(x_0)|<1$
- 3) –°—á–∏—Ç–∞–µ–º $x_n=f(x_{n-1})$
- 4) –°—á–∏—Ç–∞–µ–º –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ $\varepsilon>|x_n-x_{n-1}|$;
- 5) –î–æ—Å—Ç–∏–≥–∞–µ–º –∫—Ä–∏—Ç–µ—Ä–∏—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∏. –ó–∞–ø–∏—Å—ã–≤–∞–µ–º $x_n$ –∫–∞–∫ –æ—Ç–≤–µ—Ç.
- 6) –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –º–æ–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –∫–∞–∫ $\varepsilon>|x_n-x_{n-1}|=>\Delta$;


# –ú–µ—Ç–æ–¥ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã—Ö (–ú–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞).
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–∏–¥–∞ $f(x)=0$. –ü—É—Å—Ç—å –∫–æ—Ä–µ–Ω—å —É—Ä–∞–≤–Ω–µ–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω –∏ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[a,b]$. –¢—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ —ç—Ç–æ—Ç –∫–æ—Ä–µ–Ω—å —Å –Ω–∞–ø–µ—Ä–µ–¥ –∑–∞–¥–∞–Ω–Ω–æ–π —Ç–æ—á–Ω–æ—Å—Ç—å—é $\varepsilon$. –£—Ç–æ—á–Ω–∏–º –∫–æ—Ä–µ–Ω—å –º–µ—Ç–æ–¥–æ–º –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã—Ö. <br/>
–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Å–º—ã—Å–ª–º–µ—Ç–æ–¥–∞ –∫–∞—Å–µ—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –Ω–∏–∂–µ

–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ $[a,b]$ –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $y=f(x)$ –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π, –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω–æ–π —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É $(a,f(a))$ –∏–ª–∏ $(b,f(b))$. –ü—Ä–∏ —ç—Ç–æ–º –Ω–∞–∫–ª–∞–¥—ã–≤–∞—é—Ç—Å—è —É—Å–ª–æ–≤–∏—è –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[a,b]$:<br/>

- 2) –§—É–Ω–∫—Ü–∏—è $f\prime(x)$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
- 3) –§—É–Ω–∫—Ü–∏—è $f\prime\prime(x)$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π –±—É–¥–µ—Ç –∏–º–µ—Ç—å –≤–∏–¥ $y-f(x_0)=f\prime(x)(x-x_0)$. <br/>
–û—Ç—Å—é–¥–∞ –ø–æ–ª—É—á–∞–µ–º —Ñ–æ—Ä–º—É–ª—É –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã X –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –æ—Å–∏ OX –∏ –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π. <br/>$x_{n+1}=x_n-\frac{f(x_n)}{f\prime(x_n)}$ <br/>
