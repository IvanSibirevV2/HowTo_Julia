```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(0,4)
_ylim=(-10, 10)
plot(title = "")
plot!(x->(x*x*x-3*x*x+12*x-12), label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
#plot!(size=(400,200))
```

```julia
E_=0.1
f=x->(x*x*x-3*x*x+12*x-12)
f_=x->(3*x*x-3*2*x+12)
Niutona=x->(x-f(x)/f_(x))
x0=2;
L"x_0=%$(x0)"|>display
x1=x0|>Niutona;
L"x_1=%$(x1)"|>display
L"E=%$(E_)>|x_0-x_1|=%$(abs(x0-x1)) - %$(E_>abs(x0-x1))"|>display
x2=x1|>Niutona;
L"x_2=%$(x2)"|>display
L"E=%$(E_)>|x_1-x_2|=%$(abs(x1-x2)) - %$(E_>abs(x1-x2))"|>display
x3=x2|>Niutona;
L"x_3=%$(x3)"|>display
L"E=%$(E_)>|x_2-x_3|=%$(abs(x2-x3)) - %$(E_>abs(x2-x3))"|>display
```
