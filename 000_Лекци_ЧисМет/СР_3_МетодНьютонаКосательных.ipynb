Шаг первый. Графически локализовать корни.

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения
_xlim=(-2.5,4)
_ylim=(-3, 3)
plot(title = "")
plot!(x->(x^3-3x^2+3), label = L"2*x^3-3*x^2-12x^1-5",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
#plot!(size=(400,200))
```

Методом половинного деления уточникть корни до заданной точности

```julia
function Sign(x::Core.Real);return Base.cmp(x,0);end;
function MetoDelenia(
    _xlast::Core.Real
    ,F=(x^3+3*x^2-24*x^1-10)
    ,F_=(3x^2+3*2*x-24)
    ,e::Core.Real=0.01
)
    i=0;
    _flag =false
    while ((!_flag)&&(i<10));
        i=i+1;
        _xnext=_xlast-F(_xlast)/F_(_xlast)
        F(_xlast)|>println;
        F_(_xlast)|>println;
        _flag=Base.abs(_xlast-_xnext)<e
        "$_xnext $_flag"|>println;
        _xlast=_xnext;
    end;
end;
#MetoDelenia(1,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001);
#""|>println;
#MetoDelenia(4,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001)
#""|>println;
#MetoDelenia(-2,x->(x^3+3*x^2-24*x^1-10),x->(3x^2+3*2*x-24),0.000001)
MetoDelenia(3,x->(x^3-3x^2+3),x->(3*x^2-2*3x),0.1);
```
