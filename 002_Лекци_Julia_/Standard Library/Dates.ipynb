# Dates
Используется для работы с **датой и временем**.

```julia
2013|>Dates.DateTime|>typeof|>println
2013|>Dates.DateTime|>typeof|>typeof|>println
fieldnames(Dates.DateTime)|>println
Dates.DateTime(2013)|>println
Dates.DateTime(2013,7)|>println
Dates.DateTime(2013,7,1)|>println
Dates.DateTime(2013,7,1,12)|>println
Dates.DateTime(2013,7,1,12,30)|>println
Dates.DateTime(2013,7,1,12,30,59)|>println
Dates.DateTime(2013,7,1,12,30,59,1)|>println
```

```julia
fieldnames(Dates.Date)|>println
Dates.Date(2013)|>println
Dates.Date(2013,7)|>println
Dates.Date(2013,7,1)|>println
Dates.Date(Dates.Year(2013),Dates.Month(7),Dates.Day(1))|>println
Dates.Date(Dates.Month(7),Dates.Year(2013))|>println
```

```julia
DateFormat("y-m-d")|>x->(
    Date("2015-01-01",x)|>println;
    Date("2015-01-02",x)|>println;
)
```
