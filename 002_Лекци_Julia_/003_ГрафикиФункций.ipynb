При помощи plot! можно добросить к предыдущему графику временной ряд. Это следующие 2 примера

```julia
x = 1:10;
y = rand(10, 2); # 2 columns means two lines
z = rand(10);
p = plot(x, y)
p|>display
plot!(x, z)|>display
```

```julia
x = 1:10;
y = rand(10, 2); # 2 columns means two lines
z = rand(10);
p = plot(x, y)
p|>display
plot!(p,x, z)|>display
"#"^80|>display;
```

plot обладает диким колличеством параметров, все их изучить есть возможность только по примерам из интернета.<br/>
- plot - это объект графика, очень похож на паттерн одиночка (Singleton). Singleton может существовать в единстенном экземпляре, создание второго экземпляра переписывает предыдущий.
- plot! - позволяет добавить занчение атрибута к ранее созданному графику
- ylabel! - можно докинуть атрубут в уже ранее созданному графику
- display - централизованный метод отображения объектов,Постфактум (после их создания).
- если ничего не выводить, то отображается последний возвращенный объект, то есть последни упомянутый. А это plot.
- #########################
- title - парамерт подписи над графиком
- label - параметр легенды карты. Как видете он может принимать шаблоны латеха, только в них равно отображается плохо.
- xlabel,ylabel - параметры подписей вдоль осей

```julia
plot(x->(pi*sin(x)), title = "Pi Sinys", label = L"\pi*\sin(x)")
plot!(x->0, label = "y=0")
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#Пример демонстрирует что у каждого графика можно сделать свой диапазон отображения

plot(title = "Sinys")
plot!(x->(pi*sin(x)), label = L"\pi*\sin(x)",xlim=(-2.5,2.5),ylim=(-4, 4))
plot!(x->(pi*cos(x)), label = L"\pi*\cos(x)",xlim=(-3.5,3.5),ylim=(-4, 4))
plot!(x->0, xlim=(-7,7), ylim=(-4, 4))
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

```julia
using Plots;
using Latexify,LaTeXStrings,SymEngine;
#################################################################
_xlim=(-7,7)
_ylim=(-4, 4)
plot(title = "Sinys")
plot!(x->(pi*sin(x)), label = L"\pi*\sin(x)",xlim=_xlim,ylim=_ylim)
plot!(x->(pi*cos(x)), label = L"\pi*\cos(x)",xlim=_xlim,ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```

```julia
plot(title = L"x^3+x^2+x+5")
_xlim=(-1.882,-1.881)
_ylim=(-0.01, 0.01)
plot!(x->(x^3+x^2+x+5), label = L"x^3+x^2+x+5",xlim=_xlim, ylim=_ylim)
plot!(x->0, xlim=_xlim, ylim=_ylim)
plot!(xlabel = "OX")
ylabel!("OY")
plot!(size=(400,200))
```
