# Операции над матрицами.

Наченем с двух способов вывода объектов в Jupiter/Julia

```julia
"(^_^)"|>print;"_/"|>println;
```

```julia
"(^_^)_/"|>display
```

Теперь постепенно разберемся с массивами и матрицами

```julia
[1 2 3 4]|>display
[1,2,3,4]|>display
[1,2,3,4]'|>display
```

```julia
[1 2 ; 3 4]|>display
[1 2
3 4]|>display
```

Между println и display есть разница. display обладает приоритетом над println, и по тому отображается раньше.<br/> 

```julia
A = [1 2 ; 3 4]
A|>println
A|>display
```

```julia
A = [1 2 ; 3 4]
B = [5 6;7 8]
A|>display
B|>display
A+B|>display
```

```julia
#По элементное сложение
A .+ B
```

А теперь вопрос на засыпку. Следующие 2 примере в чем разница:

```julia
[1 1 ; 1 1] .+ [-1, 1]
```

```julia
[1 1 ; 1 1] .+ [-1 1]
```

Теперь пришло время перемножения матриц. Следующие два примера, в чем разница?


```julia
[1 2 ; 3 4] * [5 6 ; 7 8]
```

```julia
#По элементное умножение
[1 2 ; 3 4] .* [5 6 ; 7 8]
```

```julia
#Пример, обратные матрицы
A = [2 3 2; 1 2 -3; 3 4 1]
A|>display
A^(-1)|>display
A*A^(-1)|>display
"Все верно"|>display
```

Однозначно есть и более простые способы вывести матрицу на экран, но ценность этого примера в том что он отвечает на все вопросы:
- Как получить строку/столбец матрици
- Как узнать размерность матрици
- Как получить и изменить отдельный параметр матрици

```julia
#Задаем матрицу по строкам
A = [2 3 2; 1 2 -3; 3 4 1; 9 7 5]
A|>size|>display
A[2,2]=99
for i=1:size(A)[1]
    for j=1:size(A)[2]
        lpad(A[i,j],3)|>print
    end;
    ""|>println;
end;
```

size(A) - функция получения размера массива, возвращает Core.Tuple.

```julia
A = [2 3 2; 1 2 -3; 3 4 1]
A|>size|>typeof|>display
A|>size|>display
A|>length|>display
"#"^20|>display
A = [ 0 0 1 0 0]
A|>display
A|>typeof|>display
A|>size|>display
```

И на последок...
- length(A) - число элементов.
- ndims(A) - число размерностей.
- size(A) - кортеж размерностей.
- size(A, n) - размерность в заданном направлении.
- copy(A) - создание копии массива.

Последняя функция крайне полезная, только не переусерствуйте. copy в коонечном итоге может занять до трети времени вычислительного алгоритма.

- ndims(A) - число размерностей.
- size(A) - кортеж размерностей.

```julia
A = [2 3 2; 1 2 -3; 3 4 1; 9 7 5]
A|>length|>println
A|>ndims|>println
A|>size|>println
size(A,2)|>println
copy(A)
```

```julia
A = [2 3 2; 1 2 -3; 3 4 1; 9 7 5]
A[:,:]|>display
A[1:2,:]|>display
A[3:3,:]|>display
A[:,3:3]|>display
```

# И снова массивы
##############################################
Степень проработки низкая (первый отжим)

```julia
a = Int64[]
push!(a,1);push!(a,2);push!(a,4);push!(a,3);
a|>display
pop!(a);pop!(a);
a|>display
push!(a,5);
```

```julia
a = Int64[]
push!(a,1);push!(a,2);push!(a,3);push!(a,4);
"$(a[1])...$(a[end])"|>display
a|>display;
circshift(a,2)|>display;
circshift(a,-2)|>display;
############################
```

```julia
a = Int64[]
push!(a,4);push!(a,3);push!(a,2);push!(a,1);
a|>sort|>display;
a|>display;
```

```julia
#Удаляем элементы по индексу
arr = [3,4,5]
splice!(arr,2)|>display;
arr|>display;
```

```julia
#Объединяем списки
a = [3,4,5]
a|>display;
b = [1,2,3]
append!(a,b)|>display;
```

```julia
#Словарь
D = Dict('a'=>2, 'b'=>3)
D|>display;
haskey(D, 'a')|>display;
haskey(D, 'c')|>display;
```

```julia
#Возвращает индекс если такой элемент есть, иначе возвращает указанное число
D = Dict('a'=>2, 'b'=>3)
get(D,'a',9)|>display;
get(D,'b',9)|>display;
get(D,'R',9)|>display;
```
