```julia
#через Julia#import Pkg; Pkg.add("Interact")
using Plots
X = 1:10; Y = rand(10); # These are the plotting data 
plot(X,Y, label="my label")
```

```julia
X = [1 2 3 4]'
#Y=X|>x->((4)*x+6)
Y=[4 4 7 2]'
plot(Y, label="my label")
```

```julia
plot(x->(x), label="y=x")
plot!(x->(-x), label="y=-x")
plot!(x->(3*x+2), label="y=3*x+2")
plot!(x->(10*sin(x)), label="y=10*sin(x)")
plot!(x->(10*cos(x)), label="y=10*cos(x)")
```

```julia
plot(x->(10*tan(x)), label="y=10*tan(x)")
```

```julia
plot(x->(10*atan(x)), label="y=10*(x)")
```

```julia
using LinearAlgebra, Plots, Interact
x = 1:10; y = rand(10); # These are the plotting data 
plot(x,y, label="my label")
```
