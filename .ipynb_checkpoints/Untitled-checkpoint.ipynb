Труба - это специфический способ записи вызова функции:<b> [параметр][труба][лямбда выражение] </b>

Что равносильно:<b> [лямбда выражение] ([параметр]) </b>

В документации это называется как "Операции, применимые к функциям". Мы же будем назыать трубой (В стиле  ∗
 .bat файлов). Этот прием позволяет сэкономить время принаписании программных кодов. Не придется бегать в начало и в конец строки для постановки скобок.

```julia
#Обычный привет мир
println("HelloWorld!")
#Привет мир через трубу
"Hello, World!"|>println
#Точка остановки программы через чтения с экрана
readline()
```

Имеет смысл ассоциировать файлы $*$.jl файл с программой julia.exe. Достаточно один раз указать, в какой программе запускать файл по умолчанию

```julia
#Как создать функцию.#Спойлер есть еще сдесяток способов, но этот самый удобный
function Writer1()
    println("Writer1")
end
Writer1()
#...
function Writer2()println("Writer2");end;Writer2();
#...
Writer3() = println("Writer3");Writer3()
#...
Writer4() = (println("1 - Writer4");println("2 - Writer4"););Writer4();
#...
Writer5() = begin println("Writer5"); end;Writer5();
#...
Writer6=()->println("Writer6");Writer6();
#...
Writer7=()->begin println("Writer7");println("Writer7");end; Writer7()
#...
Writer8=()->(println("Writer8");println("Writer8"););Writer8();
```

```julia
#Пример формы записи сокращенной и полной формы лямбда выражений
"Writer1"|>println
"Writer2"|>x->println
"Writer3"|>x->println(x)
"Writer4"|>(x)->println(x)
"Writer5"|>(x::Core.String)->println(x)
"Writer6"|>(x::Core.String)->(println(x);)
```

```julia
#Пример о том, что возвращается последнее упомянутое значение... return необязательный
function bar()
    x = "Hello, Mars!"; return x;
end;
bar()|>println
function bar()
    x = "Hello, Mars!";
end;
bar()|>println
```

```julia
#Пример самовозвращаемой функции
function w() 
    println("w");
    w;
end;
w()()()()()()()()
```

```julia
function bar(x) print(x,";"); x=x+1; end;
5|>bar|>bar|>bar|>bar|>bar
println();
bar(bar(bar(bar(bar(5)))))
```

Также труба позволяет оперировать локальными переменными не создавая их. Создается лямбда контекст существования переменной.

```julia
4|>x->(
    x|>println;
    x+2|>println;
    (x-32)|>y->(
        y|>println;
        (y+4)|>println;
    )
)
```

Такой подход несет массу плюсов при отладке и разработке высокоуровневых программных кодов.<br/>
Нет необходимости заморачиваться над названиям переменных.<br/>
Все операции что над ними происходят можно дописать по контексту.<br/>
Единственный минус проявляется только тогда когда вы становитись высоко квалифицированным разработциком и начинаете в программном коде обращать внимание на скорость исполнения программного кода. В таком случае вы знаете что каждый вызов функции занимает время примерно сравнимое, большее в 10 раз чем сложение двух переменных.<br/>
Последний тезис имеет значение только в оптимизации низкоуровневых часто повторяемых операций. На этом этапе речь может зайти о написании императивных моноблоков кода в переделах одной функйии, об алгоритмической оптимизации и написании функций в виде моноблоков ассемблерных вставок. На этом этапе имеет смысл прежде чем кидаться в дебри обратить внимание на аппарат макросов. По идее функция макроса вставляет вместо себя свой программный код, чем и икономит время исполнения по сравненнию с вызовом функции. Последние работает в случае полной компиляции приложения. В случае интерпритации - макросы остаются под вопросом, это предмет экспериментов.
Так что массовое использование труб и лямбда выражений не повредит производительности на участках программного кода, далеих от аппаратного кода и часто вызываемых функций, то есть почти на всех учатсках кода.
